# Stretching Contrast of an Image


```python
img3=input("Enter image")
img4=input("Output image")
```

```python
k=plt.imread(img3)
```

```python
max1=np.max(k)
min1=np.min(k)
```

```python
stretched = ((k-min1)/(max1-min1))*255
clipped=np.clip(stretched,0,255)
type_clip=clipped.astype(np.uint8)
```

```python
plt.imshow(k,cmap='gray')
plt.show()

plt.imshow(type_clip,cmap='gray')
plt.show()
```
