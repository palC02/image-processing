#### To read an image(RGB) and then extract all pixels from the image and store in a 3D array.
#### (i)   Display RGB image on screen
#### (ii)  Add a border around the image
#### (iii) Multiply and arrange them in a single image

```python
#Importing modules

import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading data from file
path = r'/content/drive/MyDrive/Publicly Shared/MSc Computer Science SXCCAL/M.Sc. Sem 4/Img Proc (AN Sir)/20250121/'

file1 = input("Enter any RGB image file name: ")
file2 = input("Enter the output image file name: ")

data1 = plt.imread(path+file1)  # To extract all pixels in data1
```

```python
#Assigning shape and extracting Red, Green and Blue components

shape1 = data1.shape

r,c,d = shape1[0],shape1[1],shape1[2]

print("Rows = ",r,"    Columns = ",c,"    Depth = ",d)

#To extract the Red, Green and Blue components separately
R, G, B = data1[:,:,0], data1[:,:,1], data1[:,:,2]
```

```python
#Adding a border around the image

data2 = np.copy(data1)

data2[:,0:20,:] = 0
data2[:,c-20:c,:] = 0
data2[0:20,:,:] = 0
data2[r-20:r,:,:] = 0
```

```python
#Setting the dimensions of the final image

n = int(input("Enter the no. of times to copy the image along x-axis: "))
m = int(input("Enter the no. of times to copy the image along y-axis: "))

shape3 = (m*r, n*c, d)
data12 = np.zeros(shape3, dtype = np.int16)
r3, c3, d3 = shape3[0], shape3[1], shape3[2]
print('Final image rows: ', r3,'\t Final image columns: ', c3,'\t Final image depth: ', d)
```

```python
#Multiplication

data12[0:r,0:c,:] = data2[:,:,:]

for i in range(1,m):
  data12[i*r:(i+1)*r,0:c,:] = data2[0:r,0:c,:]

for j in range(1,n):
  data12[0:r3,j*c:(j+1)*c,:] = data12[0:r3,0:c,:]
```

```python
#To display all images in a single canvas

rows = 3
cols = 1

fig = plt.figure(figsize=(10, 24))

#Original Image:
fig.add_subplot(rows, cols, 1)
plt.axis('off')
plt.title('Original Image')
plt.imshow(data1)

#Bordered Image:
fig.add_subplot(rows, cols, 2)
plt.axis('off')
plt.title('Bordered Image')
plt.imshow(data2)

#Multiplied Image:
fig.add_subplot(rows, cols, 3)
plt.axis('off')
plt.title('Multiplied Image')
plt.imshow(data12)

plt.savefig(path+file2)
```
